```python
#Create PIE CHART

car_make_count = df['Car Make'].value_counts()
car_make_count


fig, ax = plt.subplots(figsize=(4,4))
ax.pie(car_make_count, labels=car_make_count.index, autopct='%1.1f%%', startangle=90)
ax.axis('equal')
ax.set_title('Distribution of Car Makes')
plt.show()
```

```python
#Create PIE CHART

car_models = df['Car Model'].value_counts()
car_models


fig, ax = plt.subplots(figsize=(4,4))
ax.pie(car_models, labels=car_models.index, autopct='%1.1f%%', startangle=90)
ax.axis('equal')
ax.set_title('Distribution of Car Models')
plt.show()
```
