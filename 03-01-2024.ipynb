```python
import numpy as np

a = np.array([1,2,3])
b = np.array([(1,2,3),(4,5,6)])

print("A :",a)
print("B :",b)
```

```python
a = np.array([1, 2, 3])
b = np.array([(1,2,3), (4,5,6)])

print("----------")
print("A :",a)
print("A's ndim",a.ndim)
print("A's size",a.size)
print("A's shape",a.shape)
print("A's dtype",a.dtype)
print("A's itemsize",a.itemsize)
print("----------")
print("B :",b)
print("B's ndim {}".format(b.ndim))
print("B's size",b.size)
print("B's shape",b.shape)
print("B's dtype",b.dtype)
print("B's itemsize",b.itemsize)
```

```python
raj = np.zeros((2,3))
print("Zeros Array : {}".format(raj))
print("\n")
b = np.ones((2,3))
print("Ones Array :",b)
print("\n")
c = np.empty((2,3))
print("Empty Array :",c)
print("\n")
d = np.arange(1,2,0.2)
print("Arange Array :",d)
print("\n")
e = np.linspace(1,3,7)
print("Linespace Array :",e)
print("\n")
f = np.random.random((2,3))
print("Random Array :",f)
```
