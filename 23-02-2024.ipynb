```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'x':range(1,11),'y':[5,7,3,8,10,6,2,4,9,3]}
print(data)
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
#Line Chart

plt.plot(df['x'],df['y'],marker='o',linestyle='dashed')
plt.show()
```

```python
#Line Chart


plt.plot(df['x'],df['y'],marker='^')
plt.title("LineChart")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
```

```python
data = {'x':['A','B','C','D','E'],'y':[15,30,25,10,20]}
print(data)

print("\n")

df = pd.DataFrame(data)
print(df)

print("\n")

#Vertical Bar Chart
plt.bar(df['x'],df['y'])
plt.title("Bar Chart")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(axis='y')
plt.show()
```

```python
data = {'x':['A','B','C','D','E'],'y':[15,30,25,10,20]}
print(data)

print("\n")

df = pd.DataFrame(data)
print(df)

print("\n")

#Vertical Bar Chart
plt.barh(df['x'],df['y'])
plt.title("Bar Chart")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(axis='y')
plt.show()
```

```python
#Scatter Plot

plt.scatter(df['x'],df['y'],linewidths=15,marker=8)
plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
```

```python
#PIE Chart

data = {'labels':['A','B','C','D','E'],'sizes':[15,30,25,10,20]}

df = pd.DataFrame(data)
print(df)

print("\n")

plt.pie(df['sizes'],labels=df['labels'],autopct='%1.1f%%',startangle=90)
plt.title("PIE CHART")
plt.show()
```

```python
#Histogram Chart

data = {'values':np.random.randn(1000)}
df = pd.DataFrame(data)
print(df)

print("\n")


plt.hist(df['values'],edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Histogram Chart

data = {'values':np.random.randn(1000)}
df = pd.DataFrame(data)
print(df)

print("\n")


plt.hist(df['values'],bins=10,edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
